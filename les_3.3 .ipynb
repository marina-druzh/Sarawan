```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Урок 3. Задание 3.2 построения окружности в полярных координатах
x = np.linspace(-np.pi/2, np.pi/2, 100)
y = 8*np.cos(x)
plt.polar(x, y)
```

```python
#Вариант построения окружности в полярных координатах
fig = plt.figure()
phi = np.linspace(0, 2*np.pi, 100)
r = np.linspace(5, 5, 100)
plt.polar(phi, r, color = 'm')
plt.show()
```

```python
# Задание 3.4 - Решение систем уравнений
x = np.linspace(-2, 3, 201)
plt.plot(x, (np.exp(x) - 1 + x)/x)
plt.plot(x, x**2 - 1)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-2,8) 
plt.grid(True)
plt.show()

from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return ( x**2 - 1-y, (np.exp(x) - 1 + x)/x - y)
            
x1, y1 =  fsolve(equations, (-2, 1))
x2, y2 =  fsolve(equations, (2, 6))
              
print (x1, y1)
print (x2, y2)
```
