```python
# График окружности в декартовых координатах 
fig = plt.figure(figsize=(5., 5.))
phi = np.linspace(0, 2*np.pi, 100)
x = 10*np.cos(phi)
y = 10*np.sin(phi)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.plot(x, y)
plt.show()
```

```python
"""Построение гиперболы"""
x = np.linspace(-50.1, 50.1, 1000)
a = 7
b = 3
y = [pow(i**2*b**2/a**2 - b**2, 0.5) for i in x]
y1 = [-pow(i**2*b**2/a**2 - b**2, 0.5) for i in x]

a1 = 3
b1 = 5
y2 = [pow(i**2*b1**2/a1**2 - b1**2, 0.5) for i in x]
y3 = [-pow(i**2*b1**2/a1**2 - b1**2, 0.5) for i in x]

plt.plot(x, y, color='m')
plt.plot(x, y1, color='m')
plt.plot(x, y2, color='y')
plt.plot(x, y3, color='y')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
# задание 2.5 - Построение графика 2-х параллельных плоскостей
from pylab import *
from mpl_toolkits.mplot3d import Axes3D

A = 2
B = 3
D1 = 5
D2 = 25


fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 2)
Y = np.arange(-50, 50, 20)
X, Y = np.meshgrid(X, Y)

Z = A*X + B*Y + D1
Z2 = A*X + B*Y + D2


ax.plot_surface(X, Y, Z)
ax.plot_surface(X, Y, Z2)
ax.scatter(0, 0, 0,'z',50,'red')
show()
```

```python
# Задание 2.5 - Постороение 3-х мерного графика поверхностей 2-го порядка - двуполостный гиперболоид
a = 4
b = 2
c = 6

fig = figure()
ax = Axes3D(fig)
X = np.arange(-50, 50, 2)
Y = np.arange(-50, 50, 2)
X, Y = np.meshgrid(X, Y)

Z =pow((X**2/a**2 + Y**2/b**2 + 1), 0.5)*c
Z1 =-pow((X**2/a**2 + Y**2/b**2 + 1), 0.5)*c
ax.plot_surface(X, Y, Z)
ax.plot_surface(X, Y, Z1)
ax.scatter(0, 0, 0,'z',50,'red')
show()
```

```python
# Задание 2.5 - Постороение 3-х мерного графика поверхностей 2-го порядка  - гиперболический параболоид
p = -4
q = 6

fig = figure()
ax = Axes3D(fig)
X = np.arange(-50, 50, 2)
Y = np.arange(-50, 50, 2)
X, Y = np.meshgrid(X, Y)

Z = (X**2/p + Y**2/q)/2


ax.plot_surface(X, Y, Z)
ax.scatter(0, 0, 0,'z',50,'red')
show()
```
